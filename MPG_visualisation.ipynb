import seaborn

```python
import seaborn as sns
```

get names of seaborn inbuilt datasets

```python
sns.get_dataset_names()
```

load miles per gallon dataset

```python
sns.load_dataset("mpg")
```

create dataset

```python
df = sns.load_dataset("mpg")
```

creating histplot of weight column

```python
sns.histplot(data=df,x="weight")
```

most common weight is between 2000 and 2500.

create kdeplot for mpg column

```python
sns.kdeplot(data=df,x="mpg")
```

most cars have mpg between 15 and 35

scatterplot between mpg and weight. looking for correlation.

```python
sns.scatterplot(data=df,y="mpg",x="weight")
```

Negative correlation between mpg and weight. The higher the weight the lower the mpg

creating scatterlplot comparing mpg and model year. 

```python
sns.scatterplot(data=df,y="mpg",x="model_year")
```

Over time cars mpg got better compared to old years

creating scatterplot between mpg and displacement

```python
sns.scatterplot(data=df,y="mpg",x="displacement")
```

negative correlation. As displacement increases mpg decreases. 

creating boxplot to compare mpg and country of origin. trying to find which country has best mpg generally.

```python
sns.boxplot(data=df,y="mpg",x="origin")
```

Cars of Japanese origin have the best mpg

creating scatterplot with hue for colour encoding to differentiate between country of origin. comparing model year and mpg.

```python
sns.scatterplot(data=df,y="mpg",x="model_year",hue="origin")
```

Japan has consistently made the cars with best mpg. Sometimes Europe takes over, but America is always at the bottom.

creating scatterplot between mpg and horsepower.

```python
sns.scatterplot(data=df,y="mpg",x="horsepower")
```

Negative correlation. The higher the horesepower the lower the mpg. Higher horsepower costs more fuel. 

creating scatterplot between mpg and cylinders

```python
sns.scatterplot(data=df,y="mpg",x="cylinders")
```

4 cylinder engines have the most mpg. the more cylinders the more powerful the engline, but they consume more fuel and lead to lower mpg. 

creating scatterplot between mpg and acceleration 

```python
sns.scatterplot(data=df,y="mpg",x="acceleration")
```

mpg and acceleration have no correlation.
